## Functions declared

```python
def displaydf(dataframe, cols = None, rows = 10):
    with pd.option_context("display.max_columns", cols):
        with pd.option_context("display.max_rows", rows):
            display(dataframe)
    return True
```

## Importing Dataset

```python
accidents = pd.read_csv('../data/accidents_bcn_clean.csv')
accidents_causes = pd.read_csv('../data/accidents_bcn_causa_cleaned.csv')
accidents_gender = pd.read_csv('../data/accidents_bcn_gender_cleaned.csv')
```

```python
for i in [accidents_causes,accidents_gender]:
    accidents = accidents.merge(i, on='Numero_expedient')
```

```python
accidents = accidents.drop(columns=['Numero_expedient','Descripcio_tipus_dia', 'Numero_morts', 'Numero_lesionats_lleus', 'Numero_lesionats_greus','Descripcio_torn'])
```
